```python
import cv2
import numpy as np
import os
```

```python
path = "front_face.xml" 
faceCascade = cv2.CascadeClassifier(path)
for folder in os.listdir("Original Images"):
    path = os.listdir("Original Images/" + str(folder))
    count = len(path)
    ptr = 0
    pointer = 0
    train_count = int(count*0.9)
    for file in os.listdir("Original Images/" + str(folder)):
        image_path = "Original Images/" + str(folder) + "/" + str(file)
        print(image_path)
        img = cv2.imread(image_path)
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = faceCascade.detectMultiScale(
            gray,     
            scaleFactor=1.2,
            minNeighbors=5,     
            minSize=(20, 20)
        )
        for (x,y,w,h) in faces:
            roi_gray = gray[y:y+h, x:x+w]
            roi_color = img[y:y+h, x:x+w]  
            if(pointer < train_count):
                cv2.imwrite('Train/' + folder + '/img_' + str(ptr) + '.jpg', roi_color)
            else:
                cv2.imwrite('Test/' + folder + '/img_' + str(ptr) + '.jpg', roi_color)
            ptr += 1
        pointer += 1
```
